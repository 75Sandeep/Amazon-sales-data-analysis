# identifying best selling products

```python
bsp = df.groupby('Item Type')['Units Sold'].sum().sort_values(ascending=False)
```

```python
bsp
```


# visualization of sales trend

```python
plt.figure(figsize=(100, 70))
sns.barplot(x='Country', y='Total Revenue', data = df)
plt.title('Sales Trend Over Time')
plt.show()
```

# BSP Visualization

```python
plt.figure(figsize=(20, 10))
sns.barplot(x=bsp.index, y=bsp.values)
plt.title('Best-Selling Products')
plt.xlabel('Item Type')
plt.ylabel('Units Sold')
plt.xticks(rotation=50)
plt.show()
```

# Maximum Revenue countrywise

```python
Max_revenue = df.groupby('Country')['Total Revenue'].sum().sort_values(ascending=False)
```

```python
Max_revenue
```

```python
plt.figure(figsize=(20, 10))
sns.barplot(x=Max_revenue.index, y=Max_revenue.values)
plt.title('Max Revenue')
plt.xlabel('Country	')
plt.ylabel('Total Revenue')
plt.xticks(rotation=50)
plt.show()
```
